```python
import time
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def hanoi(n, origen, destino, auxiliar):
    global contador_movimientos
    if n == 1:
        print("Mover disco 1 desde", origen, "hacia", destino)
        contador_movimientos += 1
        return
    hanoi(n-1, origen, auxiliar, destino)
    print("Mover disco", n, "desde", origen, "hacia", destino)
    contador_movimientos += 1
    hanoi(n-1, auxiliar, destino, origen)

# Ejemplo de uso:
origen = 'A'
destino = 'C'
auxiliar = 'B'

serie_tiempo = []  # Mover la declaración fuera del bucle
serie_movimientos = [] # Recolecta la cantidad de movimientos realizados por iteración
for i in range(1, 10):
    tiempo = []
    movimientos = []
    for n in range(1, 15):
        num_discos = n
        contador_movimientos = 0

        inicio_cronometro = time.time()
        hanoi(num_discos, origen, destino, auxiliar)
        fin_cronometro = time.time()
        tiempo_ejecucion = fin_cronometro - inicio_cronometro
        tiempo.append(tiempo_ejecucion)
        movimientos.append(contador_movimientos)

        print("Movimientos realizados:", contador_movimientos)
        print("Tiempo de ejecución para la última iteración:", tiempo_ejecucion, "segundos")
        print()

    print("Tiempos de ejecución:", tiempo)
    serie_movimientos.append(movimientos.copy())
    serie_tiempo.append(tiempo.copy()) # Agregar una copia de la lista tiempo a serie_tiempo


```

```python

cols = ["1 Disco", "2 Discos", "3 Discos", "4 Discos", "5 Discos",
        "6 Discos", "7 Discos", "8 Discos", "9 Discos", "10 Discos",
        "11 Discos", "12 Discos", "13 Discos", "14 Discos"]

data_frame = pd.DataFrame(serie_tiempo,columns=cols)
```
